# ***Git***
    git is Version Control System.In Git, the project or file is not created,
    but it only records the changes and maintains the versions.After the changes are committed,
    they should be pushed in Git. And to make new changes and
    To get a commit from Git, you have to pull.
    The most popular remote for git is *GitHub* .
    git bash is terminal for git.

##  *Commonly used commands in the terminal :*
- **pwd** : print working directory.
- **ls**  : list files and directory.
- **cd**  : change directory.
- **mkdir <dir_name>** : make directory.
- **touch <file_name>**: make file .
- **rm <file_name>**: remove file.
- **rm-r <dir_name>**: remove directory.
- **clear** : claer terminal.
## *Commonly used commands in `Git Bash`:*
- **`$ git log`** :  It displays the list of commits with user name, date and comment
- **`$ git status`** :  Show the working tree status
- **`$ git diff <file_name>`** : changes inside the file
- **`$ git branch <branch_name>`**: Create a new branch on local to make changes or fix bugs
- **`$ git clone <adress repository>`**: download the repository from github and saved on local.
  
## *Steps to add a new repository in command line :*

1.   **`$ git init`** : Create an empty Git repository. initialize the repository; Directory converted to repository.
2.   **`$ git config`** : This command is done once for system settings in the terminal.
        -    `$ git config --global user.email <email_address>`
        -    `$ git cofig --global user,name <user_name>`
3.   **`$ git add <file_name>`** :  for the desired file to be tracked by Git
4.   **` $ git commit -m <your message>`** : Record changes to the repository.
5.   **`  $ git remote add origin <https://github.com/user_name/repo_name.git>`**:
            Add a remote for each repository and specify the name of the main branch
6. **`git push -u origin master`** : This command will commit the changes made from local to GitHub.

- The `AND` operator displays a record if all the conditions separated by AND are TRUE.                            
- The `OR` operator displays a record if any of the conditions separated by OR is TRUE.                              
- The `NOT` operator displays a record if the condition(s) is NOT TRUE.                               
- The `IN` operator is a shorthand for multiple OR conditions.
- The `BETWEEN` operator selects values within a given range. The values can be numbers, text, or dates.
- A field with a `NULL` value is a field with no value. We will have to use the `IS NULL` and `IS NOT NULL` operators instead.   
- The `LIKE` operator is used in a WHERE clause to search for a specified pattern in a column . ' % ' , ' _ '  can also be used in     combinations.
                                                         
- The `LIMIT` clause is used to specify the number of records to return.


## *SQL Constrains :*

SQL constraints are used to specify rules for data in a table.                                    
Constraints can be specified when the table is created with the *`CREATE TABLE`* statement, or after the table is created with the *`ALTER TABLE`* statement.

The following constraints are commonly used in SQL:

- *`NOT NULL`* - Ensures that a column cannot have a NULL value. for example : username in site                          

```mysql
    CREATE TABLE Persons (
        ID int NOT NULL,
        LastName varchar(50) NOT NULL,
        FirstName varchar(50) NOT NULL,
        Age int);   


    ALTER TABLE Persons
        MODIFY Age int NOT NULL;
```

- *`UNIQUE`* - Ensures that all values in a column are different . for example : username , email_adress  in site
```mysql
       CREATE TABLE Persons (
            ID int NOT NULL,
            LastName varchar(50) NOT NULL,
            FirstName varchar(50),
            Age int,
            UNIQUE (ID)
    );    


        ALTER TABLE Persons
            ADD UNIQUE (ID);
```
 to define a `UNIQUE` constraint on multiple columns :
```mysql
        CREATE TABLE Persons (
            ID int NOT NULL,
            LastName varchar(50) NOT NULL,
            FirstName varchar(50),
            Age int,
            CONSTRAINT UC_Person UNIQUE (ID,LastName)
);


            ALTER TABLE Persons
                ADD CONSTRAINT UC_Person UNIQUE (ID,LastName);
```

    To drop a UNIQUE constraint, use the following SQL:                   

```mysql
        ALTER TABLE Persons
            DROP INDEX UC_Person;
```

- *`PRIMARY KEY`* - A combination of a NOT NULL and UNIQUE. Uniquely identifies each row in a table .
```mysql
        CREATE TABLE Persons (
            ID int NOT NULL,
            LastName varchar(50) NOT NULL,
            FirstName varchar(50),
            Age int,
            PRIMARY KEY (ID)
        );
  ```
```mysql
        ALTER TABLE Persons
            ADD PRIMARY KEY (ID);
```
```mysql

```              
- *`FOREIGN KEY`* - A FOREIGN KEY is a field in one table, that refers to the PRIMARY KEY in another table.
```mysql
        CREATE TABLE Orders (
            OrderID int NOT NULL,
            OrderNumber int NOT NULL,
            PersonID int,
            PRIMARY KEY (OrderID),
            FOREIGN KEY (PersonID) REFERENCES Persons(PersonID));


            ALTER TABLE Orders
                ADD FOREIGN KEY (PersonID) REFERENCES Persons(PersonID);
```                        
- *`DEFAULT`* - Sets a default value for a column if no value is specified .
```mysql
      CREATE TABLE Persons (
            ID int NOT NULL,
            LastName varchar(50) NOT NULL,
            FirstName varchar(50),
            Age int,
            City varchar(50) DEFAULT 'Tehran'
        );



        ALTER TABLE Persons
            ALTER City SET DEFAULT 'Tehran';
```

#### *Slicing :*
We can specify a range of indexes by where to start and where to end the range .                                 
which can be shown as below :  

`list[a:b]  a <= index <b`                         
`list[:b]  0 <= index <b`                            
`list[a:]  a <= index <= -1`                          
`list[:]  index >= 0 & index <= -1`                             

for example :
> *l3 : [ 8 , 1 , 5 , 7 , 1  , 9 , 2 , 3 , 1 , 8 , 6 , 4 , 5 ]*
```python
print('l3[2:6]:', l3[2:6])                           
print('l3[5:10]:', l3[5:10])                                                   
print('l3[-8:-2]:', l3[-8:-2])
print('l3[-8:8]:', l3[-8:8])
print('l3[:5]:', l3[:5])
print('l3[8:]:', l3[8:])
print('l3[:]:', l3[:])
```
output :
> l3[2:6] : [5, 7, 1, 9]                                 
> l3[5:10] : [9, 2, 3, 1, 8]                                    
> l3[-8:-2] : [9, 2, 3, 1, 8, 6]                                     
> l3[-8:8] : [9, 2, 3]                                         
> l3[ :5] : [8, 1, 5, 7, 1]                            
> l3[8: ] : [1, 8, 6, 4, 5]                                 
> l3[ : ] : [8, 1, 5, 7, 1, 9, 2, 3, 1, 8, 6, 4, 5]                                              

#### *Steps :*
*`list[a:b:n]`* That is, from index `a` to index `b`, take `n` steps forward
for example :
> *l3 : [ 8 , 1 , 5 , 7 , 1  , 9 , 2 , 3 , 1 , 8 , 6 , 4 , 5 ]*
```python
print('l3[::2]:', l3[::2])                  # From first to end of list by step 2
print('l3[-8:8:2]:', l3[-8:8:2])            # From index -8  to index 8 of list by step 2
print('l3[::-1]:', l3[::-1])                # The list is reversed
```
output :
> l3[ : : 2] : [8, 5, 1, 2, 1, 6, 5]                                
> l3[-8:8:2] : [9, 3]                                        
> l3[ : :-1] : [5, 4, 6, 8, 1, 3, 2, 9, 1, 7, 5, 1, 8]

## **While Loops :**
With the `while` loop we can execute a set of statements as long as a condition is true.                   
The number of rotations is not specified .  
Any `for` loop can be written with a `while` loop, but vice versa is not possible in all situations             
with *`break`* We can stop the loop before it goes through all the items

```python
while condition :
    statment1
    statment2
    statment3


Example :

x = 4
while x > 0:
    print('x:', x)
    x = x - 1
print('End of Loop!')
```

> x: 4

x: 1
